```python
# 2) Oscillation parameters
amplitude = 0.03    # meters peak‐to‐zero (so ±5 cm)
frequency = 0.1     # Hz (one full up‐and‐down every 10 s)
dt = 0.02           # control loop period (s) → 50 Hz
```

```python
start_t = time.time()
while True:
    t = time.time() - start_t
    # z(t)=A·sin(2πft), so v_z = A·2πf·cos(2πft)
    vz = amplitude * 2 * np.pi * frequency * np.cos(2 * np.pi * frequency * t)
    cart_vel = np.array([0.0, 0.0, vz])

    # move_tcp expects a position delta; vel * dt gives the step size
    robot.move_tcp(cart_vel * dt)
    time.sleep(dt)
```

```python
freq = 30
dt = 1.0 / freq
time_to_execute = 1
cart_vel = np.array([0.0, 0.0, 0.0])
for i in range(time_to_execute * freq):
    robot.move_tcp(cart_vel * dt)
    time.sleep(dt)
```

```python
# ←——— TUNE THESE ———→
CONTROL_FREQ = 30.0      # control loop rate in Hz
IK_FREQ      = 10.0      # IK solve rate in Hz

DT        = 1.0 / CONTROL_FREQ   # seconds per control tick
IK_PERIOD = 1.0 / IK_FREQ        # seconds between IK solves
# ←—————————————————→

t0 = time.time()
def dummy_joystick():
    """
    Replace this stub with your real read_joystick() that returns
    np.array([dx, dy, dz]) in meters per second.
    Here we do a little up/down sine for demo.
    """
    t = time.time() - t0
    vz = 0.01 * np.sin(2*np.pi*0.5*t)
    return np.array([0.0, -vz, -vz])

robot.dynamic_tcp_control(
    read_joystick_fn=dummy_joystick,
    dt=DT,
    ik_period=IK_PERIOD
)
```
